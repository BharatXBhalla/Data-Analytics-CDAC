## Distribution Plots

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

For studying univariate we have dist plot

```python
#Distribution plots
#displot
sns.displot(tips['total_bill'],kde=True,bins=8)
```

For Bivariate PLots we can use jointplot

```python
#jointplot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
#jointplot
sns.jointplot(x='total_bill',y='tip',data=tips,kind="hist")
```

```python
#jointplot
sns.jointplot(x='total_bill',y='tip',data=tips,kind="hex")
```

```python
#jointplot
sns.jointplot(x='total_bill',y='tip',data=tips,kind="kde")
```

```python
#jointplot
sns.jointplot(x='total_bill',y='tip',data=tips,kind="reg")
```

```python
#jointplot
sns.jointplot(x='total_bill',y='tip',data=tips,kind="resid")
```

```python
#rugplot
sns.rugplot(tips['total_bill']) #this gives marginal distribution
```

```python
sns.kdeplot(data=tips, x="total_bill", multiple="stack")
```

```python
sns.kdeplot(data=tips, x="total_bill", hue="time")
```

```python
sns.kdeplot(data=tips, x="total_bill", hue="time",multiple="stack")
```

```python
#pairplot
sns.pairplot(tips,kind="kde",hue='sex') #it is pair wise relationship for every numeric varibles
```

```python
#pairplot
sns.pairplot(tips,hue='sex') #it is pair wise relationship for every numeric varibles
```

```python
#matrix plot
sns.heatmap(tips.corr(),cmap='coolwarm',annot=True)
```

```python
#Grid plot
#facetgrid
g = sns.FacetGrid(tips, col="time",  row="smoker",hue="sex")
g = g.map(sns.regplot, "total_bill","tip").add_legend()
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex")
```

```python
sns.regplot(x="total_bill",y="tip",data=tips,hue="sex")
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",col="smoker",row="day")
```
