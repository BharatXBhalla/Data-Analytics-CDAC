# Categorical Data Plot

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#bar plot n count plot
sns.barplot(x='sex',y='total_bill',data=tips) #this gives mean
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x ='sex', hue = "smoker", data = tips)
 
# Show the plot
plt.show()
```

```python
#boxplot
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips)
```

```python
#violinplot
sns.violinplot(x="sex", y="total_bill", data=tips,palette='rainbow',hue='day')
```

```python
#violinplot
sns.violinplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
#violinplot
sns.violinplot(x="sex", y="total_bill", data=tips,palette='rainbow')
```

```python
#strip plot
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
#swarmplot
sns.swarmplot(x="day", y="total_bill",hue='sex',data=tips, palette="Set1")
```

```python
#factorplot
sns.factorplot(x='sex',y='total_bill',data=tips,kind='bar')
```

```python
#factorplot
sns.factorplot(x='sex',y='total_bill',data=tips,kind='strip')
```

```python
#factorplot
sns.factorplot(x='sex',y='total_bill',data=tips,kind='violin') # General Plot 
```

```python
#relplot
sns.relplot(data=tips, x="total_bill", y="tip", hue="day",kind="scatter") #relation plot // kind can be scatter or line
```
