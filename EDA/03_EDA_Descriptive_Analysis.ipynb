### Grouping Basics

##### we usally use grouping on categorical 
##### We can use corr on numerical variables

```python
dataset["drive-wheels"].unique()
```

We want to cal avg price wrt drive-wheels


```python
df_group = dataset[["drive-wheels","price","body-style"]]
```

```python
df_group.groupby(["drive-wheels"]).mean()
```

```python
df_group.groupby(["drive-wheels"],as_index=False).mean()
```

We want avg price wrt drive-wheels,body-style ==> Pivot Table



```python
dataset.pivot_table(index="drive-wheels",values="price",columns=["body-style","engine-type"])
```

```python
pivot_table=dataset.pivot_table(index=["drive-wheels"],values="price",columns=["body-style"],)
```

```python
pivot_table.fillna(0)
```

```python
import matplotlib.pyplot as plt
pivot_table.plot()
```

```python
plt.pcolor(pivot_table)
```

```python
pivot_table.columns
```
