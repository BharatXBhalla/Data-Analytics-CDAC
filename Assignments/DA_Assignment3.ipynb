```python
from sklearn.datasets import load_iris
```

```python
import pandas as pd

dataset = load_iris()
```

```python
df = pd.DataFrame(dataset['data'],columns=dataset['feature_names'])
```

```python
df.head()
```

```python
df["labels"]=dataset["target"]
```

```python
df["labels"]=df["labels"].map({0:'setosa', 1:'versicolor', 2:'virginica'})
```

## Comparison between various species based on sepal length and width using respective visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
sns.kdeplot(df[df["labels"]=="setosa"]["sepal length (cm)"])
plt.title("Setosa")
plt.subplot(1,3,2)
sns.kdeplot(df[df["labels"]=="versicolor"]["sepal length (cm)"])
plt.title("versicolor")
plt.subplot(1,3,3)
sns.kdeplot(df[df["labels"]=="virginica"]["sepal length (cm)"])
plt.title("virginica")
plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
sns.kdeplot(df[df["labels"]=="setosa"]["sepal width (cm)"])
plt.subplot(1,3,2)
sns.kdeplot(df[df["labels"]=="versicolor"]["sepal width (cm)"])
plt.subplot(1,3,3)
sns.kdeplot(df[df["labels"]=="virginica"]["sepal width (cm)"])
```

#Perform the Bi-variate Analysis via pairplot visualization

```python
sns.pairplot(df,hue="labels")
```

## Justify Sepal Length and Sepal Width features are slightly correlated with each other

```python
sns.heatmap(df.corr(),annot=True)
```

## Visualize the categorical feature Species with all other four input variables

```python
sns.pairplot(df,kind="kde",hue="labels")
```
